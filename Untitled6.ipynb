```python
n=int(input("enter the number"))
for i in range(n+1):
    if i%3==0 and i%5==0:
        print("the number",i,"is fizzbuzz",end=" ")
    elif i%3==0:
        print("the number",i,"is fizz",end=" ")
    elif i%5==0:
        print("the number",i,"is buzz",end=" ")
print()
```
