```python
import googletrans
from googletrans import Translator
```

```python
print(googletrans.LANGUAGES)
```

```python
translator = Translator()
text = "ഹലോ"
lang = translator.detect(text).lang
print(lang)
```

```python
translated_text = translator.translate(text, src = lang, dest = 'hi')
print(translated_text)
```
